Objective

    Analyze ATM cash deposits and withdrawals over 30 days using NumPy to compute daily balance, weekly trends, and cash-flow insights.

Tasks to Complete

    1: Generate:

        1.1: Array of daily deposits (between ₹5,000 and ₹40,000)

        1.2: Array of daily withdrawals (between ₹3,000 and ₹35,000)

        1.3: Compute daily remaining cash using vectorized subtraction:

        1.4: balance = deposits - withdrawals


    2: Find:

        2.1: Day with highest deposit

        2.2: Day with highest withdrawal

        2.3: Highest positive balance day

        2.4: Most negative balance day (if exists)

        2.5: Weekly Analysis (Week 1–5 → 7 days each except last)

            2.5.1: Weekly net cash flow total

            2.5.2: Week with maximum cash flow

    3: Count:

        3.1: Days where withdrawals > deposits

        3.2: Days where balance > ₹10,000


```python
import numpy as np
```

```python
#Task 1

''' 
1: Generate:

        1.1: Array of daily deposits (between ₹5,000 and ₹40,000)

        1.2: Array of daily withdrawals (between ₹3,000 and ₹35,000)

        1.3: Compute daily remaining cash using vectorized subtraction:

                    balance = deposits - withdrawals
'''

#Task 1.1

daily_deposits = np.random.randint(5000,40000,size=30).reshape(5,6)

#Task 1.2

daily_withdrawals = np.random.randint(3000,35000,size=30).reshape(5,6)

#Task 1.3

balance = daily_deposits - daily_withdrawals

```

```python
#Task 2 

''' 
2: Find:

        2.1: Day with highest deposit

        2.2: Day with highest withdrawal

        2.3: Highest positive balance day

        2.4: Most negative balance day (if exists)

        2.5: Weekly Analysis (Week 1–5 → 7 days each except last)

                2.5.1: Weekly net cash flow total

                2.5.1: Week with maximum cash flow
'''

#Task 2.1

day_highest_depo = np.argmax(daily_deposits) +1 

#Task 2.2

day_highest_wd  = np.argmax(daily_withdrawals)+1

#Task 2.3

highest_pos_balance = np.argmax(balance)+1

#Task 2.4

most_neg_balance  = np.argmin(balance)+1

#Task 2.5


#Task 2.5.1

week_net_cash = np.sum(balance,axis=1)

#Task 2.5.2

week_max_cash_flow = np.argmax(week_net_cash) + 1



```

```python
#Task 3

'''
3: Count:

        3.1: Days where withdrawals > deposits

        3.2: Days where balance > ₹10,000
'''

#Task 3.1

days_wd_d = np.where(daily_withdrawals.flatten()>daily_deposits.flatten())[0]+1

#Task 3.2

day_bal = np.where(balance.flatten()>10000)[0]+1
```
