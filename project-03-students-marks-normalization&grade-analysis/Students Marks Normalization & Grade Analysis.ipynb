```python
import numpy as np
```

```python
#Task 1

"""
        1: Generate marks of 100 students (range 0–100 using NumPy).

"""

MARKS = np.random.randint(0,101,size=100)


#Task 2

""" 

2: Print the marks array.

"""

print(MARKS)


```

```python
#Task 3

"""


    3: Perform normalization (choose one or do both):

        3.1: Min-Max Normalization

        3.2: Z-Score Normalization

            


"""

#Task 3.1

min_max_norm = (MARKS - MARKS.min())/(MARKS.max()-MARKS.min())

#Task 3.2

z_score = (MARKS-MARKS.mean())/MARKS.std()


```

```python
#Task 4
"""

 6: Apply bonus curve: +5 marks (but cap maximum at 100 using np.clip()).

""" 

curve = np.clip(MARKS+5,0,100) 

```

```python
#Task 5


"""
Classify each student into grades:

        Range	Grade
        90–100	A
        80–89	B
        70–79	C
        60–69	D
        <60	F
"""
import pandas as pd

GRADES = pd.cut(
    MARKS,
    bins=[0,60,70,80,90,100],
    labels=["F","D","C","B","A"],
    include_lowest=True
)

```
